```python
import pandas as pd


def minValue(dictionary):

  mini = min(dictionary.keys(), key=(lambda k: dictionary[k]))
  return dictionary[mini]



def greedy(city, cityBefore, citiesVisited, dest) :
  
  file_location = "Dist between cities.xlsx"
  sheet = pd.read_excel(file_location)
  dictionary = {}
  
  if city != dest:
     
     
     #search all the cities in the second column that are neighbours of city and add them in a dictionary 
     #key: neighbour of city
     #value: dist between them  
     i = 0
     for x in sheet['first']:
    
      if x == city :
        dictionary[sheet['second'].loc[i]] = sheet['dist'].loc[i]
      
      i += 1
    
     #search all the cities in the first column that are neighbours of city and add them in a dictionary 
     #key: neighbour of city
     #value: dist between them 
     i = 0
     for x in sheet['second']:
    
      if x == city :
        dictionary[sheet['first'].loc[i]] = sheet['dist'].loc[i]
      
      i += 1
     
     #add the city to the visited cites list
      citiesVisited += [city]
      
      #if the neighgbours dictionary contains a city already visited dictionary.pop() it 
      for c in citiesVisited:
          
          if c in dictionary.keys():
            dictionary.pop(c)
     
     
     if len(dictionary) != 0:
       #get the key that has the min value (neighbour that has the minimal distance from city)
        min_dist = min(dictionary, key=dictionary.get)
        
        print(city + "-" + min_dist  + ":", minValue(dictionary))

        #if destination between the neighbours dictionary take it out print it and stop
        if dest in dictionary.keys():
          dictionary.pop(dest)
          print(dest)
          return
        else:
          #else do that again for the neighbour with minimal distance from city
          greedy(min_dist, city, citiesVisited, dest)
     
      
  else:
    citiesVisited += [city]
    print(city)
    return 

print("Path:")
greedy("Paris", "", [], "Bucharest")
```
